```python
##TASK1
1. Difference between "Love" and "love" in NLP

In NLP, "Love" (capital L) and "love" (small l) are technically treated as different words unless we normalize them.

"Love" → might appear at the start of a sentence or be part of a name (like Love Sharma).
"love" → usually used as a regular verb or noun (like I love coding).

If we don’t convert text to lowercase, the model may think they are two separate tokens, which can mess up frequency counts or meaning.
That’s why we usually apply lowercasing as a preprocessing step — to treat them as the same word.

2. What happens if stopwords are not removed?

Stopwords are common words like is, the, and, in, etc.
If we don’t remove them:

1.The dataset becomes noisy.
2.Important words get less importance.
3.Models may become slower and less efficient.
4.It can affect tasks like text classification or keyword extraction.

3. Two real-world cases where removing stopwords can be harmful

(i) Sentiment Analysis 😬
Example:
"I do not like this movie"
If we remove stopwords like "not", it becomes:
"I like this movie"

(ii) Question Answering / Chatbots 🤖
Example:
"What is the time?"
I we remove stopwords:
"time"

4. Difference between Stemming and Lemmatization

Both are used to reduce words to their base form, but they work differently:
🔹 Stemming (rough & fast)
Cuts words to a root form using simple rules
Doesn’t care if the result is a real word
Example:
running → run
studies → studi

```

```python


def preprocess_text(text):
    if not text:
        return ""

    # 1. Convert to lowercase
    text = text.lower()

    # 2. Remove URLs
    text = re.sub(r'http\S+|https\S+|www\S+', '', text)

    # 3. Remove email-like patterns
    text = re.sub(r'\S+@\S+', '', text)

    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 5. Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 6. Remove special characters
    text = re.sub(r'[^a-z\s]', '', text)

    # 7. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # 8. Remove short tokens
    KEEP_WORDS = {'no', 'not'}
    words = text.split()
    cleaned_words = [word for word in words if len(word) > 2 or word in KEEP_WORDS]

    return " ".join(cleaned_words)


#Test cases
test_cases = [
    ("Remove numbers", 'I have 2 dogs'),
    ("Remove extra spaces", 'This   is    good'),
    ("Handle repeated characters", 'soooo goooood!!!'),
    ("Convert to lowercase", 'WOW!!! This is GREAT!'),
    ("Remove short tokens", 'hi is it no not bad'),
    ("Remove URLs", 'Visit http://x.com'),
    ("Remove emails", 'mail me at a@b.com'),
    ("Combined all features", 'WOW!!! I have 2 dogs!!! Visit http://x.com soooo goooood!!!')
]


for i, (label, text) in enumerate(test_cases, 1):
    print(f"\nTest {i}/{len(test_cases)}: {label}")

    result = preprocess_text(text)

    print(f"  Input  : {repr(text)}")
    print(f"  Output : {repr(result)}")
    print("-" * 60)

```

```python
import re
from collections import Counter

def preprocess_text(text):
    if not text:
        return []
    text = text.lower()
    text = re.sub(r'http\S+|https\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    KEEP_WORDS = {'no', 'not'}
    words = text.split()
    return [word for word in words if len(word) > 2 or word in KEEP_WORDS]


def token_analytics(original, tokens):
    original_count = len(original.split())
    total_tokens   = len(tokens)
    unique_tokens  = len(set(tokens))
    avg_length     = round(sum(len(w) for w in tokens) / total_tokens, 2) if tokens else 0
    noise_removed  = original_count - total_tokens
    noise_pct      = round((noise_removed / original_count) * 100, 1) if original_count else 0
    return {
        "original_word_count" : original_count,
        "total_tokens"        : total_tokens,
        "unique_tokens"       : unique_tokens,
        "avg_token_length"    : avg_length,
        "noise_removed"       : noise_removed,
        "noise_pct"           : noise_pct,
    }


def full_pipeline(text_list):
    if not text_list:
        return {
            "tokens"          : [],
            "clean_sentences" : [],
            "analytics"       : [],
            "frequency"       : {},
            "top_10_most"     : [],
            "top_5_least"     : [],
            "summary"         : {}
        }

    all_tokens      = []
    tokens_per_sent = []
    clean_sentences = []
    analytics       = []

    # ── Per Sentence Processing ───────────────────────────────────────────────
    for text in text_list:
        tokens          = preprocess_text(text)
        clean_sentence  = " ".join(tokens)
        metrics         = token_analytics(text, tokens)

        all_tokens.extend(tokens)
        tokens_per_sent.append(tokens)
        clean_sentences.append(clean_sentence)
        analytics.append({
            "original"       : text,
            "tokens"         : tokens,
            "clean_sentence" : clean_sentence,
            "analytics"      : metrics
        })

    # ── Frequency Analysis ────────────────────────────────────────────────────
    freq         = Counter(all_tokens)
    top_10_most  = freq.most_common(10)
    top_5_least  = freq.most_common()[:-6:-1]

    # ── Summary ───────────────────────────────────────────────────────────────
    summary = {
        "total_sentences"      : len(text_list),
        "total_tokens"         : len(all_tokens),
        "unique_tokens"        : len(freq),
        "avg_tokens_per_sent"  : round(len(all_tokens) / len(text_list), 2),
        "most_noisy_sentence"  : max(analytics, key=lambda x: x["analytics"]["noise_pct"])["original"],
        "most_tokens_sentence" : max(analytics, key=lambda x: x["analytics"]["total_tokens"])["original"],
    }

    return {
        "tokens"          : all_tokens,
        "clean_sentences" : clean_sentences,
        "analytics"       : analytics,
        "frequency"       : dict(freq),
        "top_10_most"     : top_10_most,
        "top_5_least"     : top_5_least,
        "summary"         : summary
    }


# ── Display Pipeline Output ───────────────────────────────────────────────────

def display_pipeline(output):
    print("=" * 65)
    print("           TASK 6 — Full NLP Pipeline Output")
    print("=" * 65)

    # Per sentence results
    print("\n📋 PER SENTENCE RESULTS")
    print("=" * 65)
    for i, entry in enumerate(output["analytics"], 1):
        m = entry["analytics"]
        print(f"\n🔹 Sentence {i}: {entry['original']}")
        print(f"   Tokens          : {entry['tokens']}")
        print(f"   Clean Sentence  : {entry['clean_sentence']}")
        print(f"   Total Tokens    : {m['total_tokens']}")
        print(f"   Unique Tokens   : {m['unique_tokens']}")
        print(f"   Avg Token Len   : {m['avg_token_length']} chars")
        print(f"   Noise Removed   : {m['noise_removed']} words ({m['noise_pct']}%)")
        print("-" * 65)

    # All tokens combined
    print(f"\n📦 ALL TOKENS COMBINED")
    print("=" * 65)
    print(output["tokens"])

    # Clean sentences list
    print(f"\n✅ CLEAN SENTENCES LIST")
    print("=" * 65)
    for i, s in enumerate(output["clean_sentences"], 1):
        print(f"  {i}. {s}")

    # Top 10 most frequent
    print(f"\n🔝 TOP 10 MOST FREQUENT WORDS")
    print("=" * 65)
    print(f"{'Rank':<6} {'Word':<20} {'Count':>5}  Bar")
    print("-" * 65)
    for rank, (word, count) in enumerate(output["top_10_most"], 1):
        bar = "█" * (count * 5)
        print(f"{rank:<6} {word:<20} {count:>5}  {bar}")

    # Top 5 least frequent
    print(f"\n🔻 TOP 5 LEAST FREQUENT WORDS")
    print("=" * 65)
    print(f"{'Rank':<6} {'Word':<20} {'Count':>5}  Bar")
    print("-" * 65)
    for rank, (word, count) in enumerate(output["top_5_least"], 1):
        bar = "█" * (count * 5)
        print(f"{rank:<6} {word:<20} {count:>5}  {bar}")

    # Summary
    print(f"\n📊 PIPELINE SUMMARY")
    print("=" * 65)
    s = output["summary"]
    print(f"   Total Sentences       : {s['total_sentences']}")
    print(f"   Total Tokens          : {s['total_tokens']}")
    print(f"   Unique Tokens         : {s['unique_tokens']}")
    print(f"   Avg Tokens/Sentence   : {s['avg_tokens_per_sent']}")
    print(f"   Most Noisy Sentence   : {s['most_noisy_sentence']}")
    print(f"   Most Tokens Sentence  : {s['most_tokens_sentence']}")
    print("=" * 65)
    print("✅ Full Pipeline Completed!")
    print("=" * 65)


# ── Run ───────────────────────────────────────────────────────────────────────

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

output = full_pipeline(test_sentences)
display_pipeline(output)

```

```python
##TAsk 7


# ── Custom Exceptions

class EmptyInputError(Exception):
    pass

class NoTextAfterCleaningError(Exception):
    pass


# ── Validator

def validate_input(text):
    # Check None or non-string
    if text is None:
        raise TypeError("Input is None. Please provide a valid string.")
    if not isinstance(text, str):
        raise TypeError(f"Expected string, got {type(text).__name__}.")

    # Check empty string
    if text.strip() == "":
        raise EmptyInputError("Input is an empty string or only whitespace.")

    # Check only numbers
    if re.fullmatch(r'[\d\s]+', text.strip()):
        raise NoTextAfterCleaningError("Input contains only numbers — no tokens to extract.")

    # Check only emojis / symbols (no ASCII letters or digits)
    cleaned = re.sub(r'[^\w\s]', '', text)
    cleaned = re.sub(r'[\d\s]', '', cleaned)
    if not any(c.isalpha() for c in text) and len(text.strip()) > 0:
        raise NoTextAfterCleaningError("Input contains only emojis or symbols — no tokens to extract.")


# ── Preprocessor

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|https\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    KEEP_WORDS = {'no', 'not'}
    words = text.split()
    tokens = [word for word in words if len(word) > 2 or word in KEEP_WORDS]

    if not tokens:
        raise NoTextAfterCleaningError("No meaningful tokens remained after cleaning.")

    return tokens


# ── Safe Wrapper

def safe_preprocess(text):
    try:
        validate_input(text)
        tokens = preprocess_text(text)
        return {
            "status"  : "success",
            "input"   : text,
            "tokens"  : tokens,
            "cleaned" : " ".join(tokens)
        }

    except TypeError as e:
        return {"status": "error", "type": "TypeError",                 "input": text, "message": str(e)}
    except EmptyInputError as e:
        return {"status": "error", "type": "EmptyInputError",           "input": text, "message": str(e)}
    except NoTextAfterCleaningError as e:
        return {"status": "error", "type": "NoTextAfterCleaningError",  "input": text, "message": str(e)}
    except Exception as e:
        return {"status": "error", "type": "UnexpectedError",           "input": text, "message": str(e)}


# ── Full Pipeline with Error Handling

def full_pipeline(text_list):
    if not text_list:
        return {"status": "error", "message": "Input list is empty."}

    results     = []
    all_tokens  = []
    success     = 0
    failed      = 0

    for text in text_list:
        result = safe_preprocess(text)
        results.append(result)
        if result["status"] == "success":
            all_tokens.extend(result["tokens"])
            success += 1
        else:
            failed += 1

    freq        = Counter(all_tokens)
    top_10_most = freq.most_common(10)
    top_5_least = freq.most_common()[:-6:-1]

    return {
        "results"     : results,
        "all_tokens"  : all_tokens,
        "frequency"   : dict(freq),
        "top_10_most" : top_10_most,
        "top_5_least" : top_5_least,
        "summary"     : {
            "total"   : len(text_list),
            "success" : success,
            "failed"  : failed
        }
    }


# ── Display

def display_results(output):
    print("=" * 65)
    print("        TASK 7 — Error Handling Report")
    print("=" * 65)

    for i, r in enumerate(output["results"], 1):
        print(f"\n🔹 Test Case {i}: {repr(r['input'])}")
        if r["status"] == "success":
            print(f"   Status  : ✅ Success")
            print(f"   Tokens  : {r['tokens']}")
            print(f"   Cleaned : {r['cleaned']}")
        else:
            print(f"   Status  : ❌ {r['status'].upper()}")
            print(f"   Type    : {r['type']}")
            print(f"   Message : {r['message']}")
        print("-" * 65)

    s = output["summary"]
    print(f"\n📊 PIPELINE SUMMARY")
    print("=" * 65)
    print(f"   Total Inputs    : {s['total']}")
    print(f"   ✅ Successful   : {s['success']}")
    print(f"   ❌ Failed       : {s['failed']}")

    if output["top_10_most"]:
        print(f"\n🔝 TOP 10 MOST FREQUENT WORDS")
        print("-" * 65)
        for rank, (word, count) in enumerate(output["top_10_most"], 1):
            bar = "█" * (count * 4)
            print(f"   {rank:<4} {word:<20} {count:>3}  {bar}")

    if output["top_5_least"]:
        print(f"\n🔻 TOP 5 LEAST FREQUENT WORDS")
        print("-" * 65)
        for rank, (word, count) in enumerate(output["top_5_least"], 1):
            bar = "█" * (count * 4)
            print(f"   {rank:<4} {word:<20} {count:>3}  {bar}")

    print("\n" + "=" * 65)
    print("✅ Error Handling Test Completed!")
    print("=" * 65)


# ── Test Cases

test_inputs = [
    # ✅ Valid
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product",
    "Visit https://openai.com for more info",
    "I am not happy with this service",
    "Nooooo this is baaad!!!",

    # ❌ Error cases
    "",                    # Empty string
    "   ",                 # Only whitespace
    "😍😍😂🔥💯",          # Only emojis
    "1234567890",          # Only numbers
    "99 100 42 007",       # Only numbers with spaces
    None,                  # None input
    12345,                 # Non-string input
]

output = full_pipeline(test_inputs)
display_results(output)
```
